```python
sales_melted = df_sales.melt(id_vars=['id', 'item', 'category', 'department', 'store', 'store_code', 'region'], var_name='d', value_name='value_sales')

sales_melted
```

```python
salted_melted = sales_melted[sales_melted["value_sales"] != 0]
salted_melted
```

```python
df_date = pd.DataFrame({'d': [f'd_{i}' for i in range(1, 1914)]})

# Create a date range from '2011-01-29' to '2016-04-24'
date_range = pd.date_range(start='2011-01-29', end='2016-04-24', freq='D')

# Add the date range as a new column
df_date['date'] = date_range
df_date["weekyear"] = df_date['date'].dt.isocalendar().week
df_date["year"] = df_date['date'].dt.isocalendar().year

def add_zero(x):
    if x < 10:
        return f'0{x}'
    else:
        return str(x)

df_date['weekyear'] = df_date['weekyear'].apply(add_zero)
df_date['yearweek'] = df_date.apply(lambda row: str(row['year']) + str(row['weekyear']), axis=1)
df_date['weekday'] = df_date['date'].dt.day_name()
df_date.info()
```

```python
df_date
```

```python
# Juntar os datasets com base na coluna 'd'
merged_df = pd.merge(sales_melted, df_date, on='d', how='inner')

merged_df
```

```python
merged_df = merged_df[merged_df["value_sales"] != 0]
merged_df
```

```python
merged_df.info()
```

```python
merged_df = merged_df[['id','yearweek','date','value_sales','d']]

merged_df.info()
```

```python
merged_df['id_yearweek'] = merged_df.apply(lambda row: str(row['id']) + str('_')+ str(row['yearweek']), axis=1)

merged_df
```

```python
merged_df = merged_df[['id_yearweek','d','date','value_sales']]

merged_df
```

```python
merged_df = pd.merge(merged_df, df_prices[['sell_price','id_yearweek']], on='id_yearweek', how='inner')

merged_df
```

```python
merged_df = merged_df[['id_yearweek','date','value_sales','sell_price']]

merged_df
```
